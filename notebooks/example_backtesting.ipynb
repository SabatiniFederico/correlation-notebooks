```python
# Crear datos de ejemplo
dates = pd.date_range(start="2023-01-01", end="2023-12-31", freq="D")
np.random.seed(42)
strategy_returns = np.cumprod(1 + np.random.normal(0.0003, 0.01, len(dates)))
benchmark_returns = np.cumprod(1 + np.random.normal(0.0002, 0.008, len(dates)))



strategy_series = pd.Series(strategy_returns, index=dates)
benchmark_series = pd.Series(benchmark_returns, index=dates)

# Ejecutar el backtest
run_backtest(
    capital=1000,
    start_date="2023-01-01",
    strategy_series=strategy_series,
    benchmark_series=benchmark_series,
    title="Estrategia vs Benchmark 2023"
)
```
